## Well logging
Geophysical well logging is a technique that uses a tool to measure physical properties of subsurface rock formations in the oil and gas industry. It provides valuable information to geologists to identify hydrocarbon reserves, plan field development, optimize well design, and make informed decisions about a project's economic viability. It is an essential tool for geologists in the oil and gas industry.

-  Geologists use well logs to understand the subsurface geology of rock formations.
-  Well logs can provide information about lithology, porosity, permeability, and other important properties of the rock formations.
-  This information helps geologists to identify potential hydrocarbon reserves and plan the development of oil and gas fields.
-  Well logs also help in optimizing well placement and design, and making informed decisions about a project's economic viability.
-  Geophysical well logging is an essential tool for geologists in the oil and gas industry

# What kind of well-logs we will use?

-  **NPHI** (neutron porosity): It is used to estimate the amount of fluids (usually hydrocarbons) contained in the rock formation by measuring the amount of neutron radiation that is emitted by the rock and reflected back to the sensor. This is important in the oil and gas industry to determine the potential productivity of a reservoir.

-  **RHOB** (bulk density): It is used to determine the weight of the rock formation per unit volume. This is important for calculating the overall density of the rock formation and understanding its mechanical properties.

-  **GR** (gamma ray): It is used to measure the amount of natural radiation that is emitted by the rock formation. This information can be used to identify certain types of rock formations, such as shale, and to estimate the amount of organic matter present in the formation.

-  **RT** (resistivity): It is used to measure the electrical resistance of the rock formation to the flow of electric current. This information can be used to determine the presence and quality of fluids within the rock formation.

-  **PEF** (photoelectric factor): It is used to measure the amount of X-ray radiation that is absorbed by the rock formation. This information can be used to identify certain types of rock formations, such as sandstone, and to estimate the amount of organic matter present in the formation.

-  **CALI** (caliper): It is used to measure the diameter of the borehole. This information is important for determining the correct size of tools to be used for further measurements and for ensuring the stability of the borehole.

-  **DT** (compressional travel time): It is used to measure the time it takes for a compressional (P-wave) sound wave to travel a known distance through the rock formation. This information can be used to determine the rock formation's mechanical properties, such as its porosity and permeability.

-  **DTS** (shear travel time): It is used to measure the time it takes for a shear (S-wave) sound wave to travel a known distance through the rock formation. This information can be used to determine the rock formation's mechanical properties, such as its elasticity and shear strength.

# LAS format

**LAS (Log ASCII Standard)** is a widely used file format in the oil and gas industry for storing well log data. It is a plain-text format that contains both header and data sections. The header section includes information such as well name, location, logging company, and units of measure. The data section includes a series of curves or logs, such as gamma ray, resistivity, and porosity, measured at different depths or intervals in the wellbore.

To read and manipulate LAS files in Python, you can use the lasio library. **Lasio** is a Python package for reading and writing LAS files. It provides a simple interface for accessing LAS file sections, curves, and metadata.

# Step 4 - Missing data

In statistics and data analysis, missing data refers to the absence of values in a dataset. There are different types of missing data, but the most common ones are Missing At Random (MAR) and Missing Not At Random (MNAR).

**Missing At Random (MAR):**
When missing data is MAR, the probability of the missing data occurring is unrelated to the value of the missing data itself, but rather it depends on other variables in the dataset. For example, if we have a dataset of students' test scores, and the data is missing for some students who were absent on the day of the exam, this missing data is MAR because it depends on the variable "attendance", which is related to the missing data.

**Missing Not At Random (MNAR):**
When missing data is MNAR, the probability of the missing data occurring is related to the value of the missing data itself. In other words, there is a systematic reason why the data is missing, and it cannot be explained by other variables in the dataset. For example, if we have a dataset of patients' medical records, and some patients refuse to disclose their medical history, the missing data is MNAR because it depends on the variable "medical history", which is related to the missing data.

**Missing Completely At Random (MCAR):**
When missing data is MCAR, the probability of the missing data occurring is completely random and unrelated to any other variables in the dataset. In other words, there is no systematic reason why the data is missing, and it is simply missing by chance. For example, if we have a dataset of survey responses, and some respondents forgot to answer some questions, the missing data is MCAR because it is completely random and unrelated to any other variables in the dataset.

**Time for a FOR loop!**

Code below creates a figure with 19 subplots arranged in a single row, where each column in the dataframe is plotted on a separate subplot. The y-axis limits are set to 4000 meters at the top and 0 meters at the bottom, with y-axis tick marks every 100 meters. The x-axis limits are set to 0 and 1. The area under the curve is filled with green color. Tick marks are removed from both axes, and y-axis tick labels are removed from all but the first subplot. A y-axis label is added to the first subplot. A legend is added with the title 'GREEN = not NaN'. The spacing between subplots is adjusted to 0, and the figure is shown.

```python
# Create a figure with 19 subplots (set it in ncols) arranged in a single row (set nrows for 1), make figsize 19x10
fig, axes = plt.subplots(nrows=1, ncols=19, figsize=(19,10))

# Iterate through each column in the dataframe and plot it on a separate subplot
for i, column in enumerate(data_df_n.columns):
    # Select the i-th subplot
    ax = axes[i]
    
    # Plot the data in the i-th column, set line width for 0
    ax.plot(data_df_n[column], data_df_n.index, lw=0)
    
    # Set the y-axis limits to 4000 meters at the top and 0 meters at the bottom
    ax.set_ylim("top", "bottom")
    
    # Set y-axis tick marks every 100 meters
    ax.set_ylim(4000, 0)
    
    # Set the x-axis limits to 0 and 1
    ax.set_xlim(0, 1)
    
    # Set the title of the subplot to the name of the column
    ax.set_title(column)
    
    # Fill the area under the curve with green color
    ax.fill_betweenx(data_df_n.index, 0, data_df_n[column], facecolor='green')
    
    # Remove tick marks from both axes
    ax.tick_params(axis='both', which='both', length=0, width=0)

    # Remove y-axis tick labels from all but the first subplot - do not correct this part
    if i > 0:
        ax.set_yticklabels([])
    
    # Remove x-axis tick labels from all subplots
    ax.set_xticklabels([])
    
    # Add a y-axis label to the first subplot
    if i == 0:
        ax.set_ylabel('DEPTH [m]', fontsize=10)
        
# Add a legend to the figure with the title 'GREEN = not NaN'
ax.legend(title='GREEN = not NaN')

# Adjust the spacing between subplots to 0
fig.subplots_adjust(wspace=0)

# Show the figure
plt.show()
```

To save a Pandas dataframe into an Excel file, you can use the **to_excel()** method provided by Pandas. The syntax for this method is as follows:


**df.to_excel('filename.xlsx', index=False)**

where df is the name of the dataframe you want to save, **'filename.xlsx'** is the name you want to give to the Excel file, and index=False specifies whether to include the index column in the saved file. If **index=True**, the index column will be included in the saved file.

You can also specify additional parameters in the **to_excel()** method, such as the sheet name and the start row and column for the data. For example:

**df.to_excel('filename.xlsx', sheet_name='Sheet1', startrow=1, startcol=2, index=False)**

This will save the dataframe to a new Excel file called **'filename.xlsx'**, with the data starting at row 2 and column 3 (since the first row and two columns are reserved for the sheet name and other information). The index=False parameter ensures that the index column is not included in the saved file.
