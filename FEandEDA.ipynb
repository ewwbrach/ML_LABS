<h4> Zbiór danych ma 82 kolumny, widać brakujące wartości, dlatego w kolejnym kroku puste wartości zostały zastąpione NaN </h4>


```python
data2 = data.replace(" ",np.NaN)
data2
```

<h4> Puste wartości zostały zamienione na NaN aby otrzymać dla nich statystkyki w kolejnym kroku </h4>

```python
miss = data2.isnull().sum()
miss_mean = data2.isnull().mean()
miss_mean

null = data2.isnull().sum().sum()
null
data2.shape

procent_pustych = null/(216*82)*100
procent_pustych

tab = []
for col in data2.columns:
    tab.append(col)

dic = {}
j=0
summ = 0
for i in tab:
    dic[i] = miss_mean[j]
    print('Braki danych dla zmiennej {}'.format(i) + ': {}'.format(round(miss_mean[j]* 100 ,2)) + "%")
    if miss_mean[j] == 0.0:
        summ = summ +1
    j = j+1

print('Liczba kolumn, które mają 0% brakujących wartości: ',summ)
```

<h4> W ten sposób otrzymano % braków danych dla poszczególnych kolumn. Wiersze kolumn, dla którym % braków danych jest ok. 5% można usunąć. Być może składniki te są trudne so wychwycenia i czujnik ich nie wychwycił lub pominięto całkowicie pomiary dla takich pierwiastków. </h4>

```python
data3 = data.replace(" ",int(-1))
for i in tab[9:]:
    data3[i] = data3[i].astype(float)
    if dic[i] <= 0.06:

        data3 = data3[data3[i] != -1]

data3 = data3.replace(-1,np.NaN)
miss_mean = data3.isnull().mean()
miss_mean
dic = {}
j=0
summ = 0
for i in tab:
    dic[i] = miss_mean[j]
    print('Braki danych dla zmiennej {}'.format(i) + ': {}'.format(round(miss_mean[j] * 100 ,2)) + "%")
    if float(miss_mean[j]) == 0.0:
        summ = summ +1
    j = j+1
print('Liczba kolumn, które mają 0% brakujących wartości po usunięciu wierszy <5%: ',summ)

```
